```python
import numpy as np
import matplotlib.pyplot as plt
from os.path import join
import random

load_dir = r'C:\Users\snehi\Documents\Python HPC\modified_swiss_dwellings'
num_to_show = 7 
building_ids_path = join(load_dir, 'building_ids.txt')

# Load list of building IDs
with open(building_ids_path, 'r') as f:
    building_ids = f.read().splitlines()

sample_ids = random.sample(building_ids, num_to_show)

fig, axes = plt.subplots(2, num_to_show, figsize=(4 * num_to_show, 8))

for idx, bid in enumerate(sample_ids):
    
    domain = np.load(join(load_dir, f"{bid}_domain.npy"))
    interior = np.load(join(load_dir, f"{bid}_interior.npy"))

    # Row 0: domain
    ax1 = axes[0, idx]
    im1 = ax1.imshow(domain, cmap='plasma')
    ax1.set_title(f"{bid} - Domain")
    ax1.axis('off')

    # Row 1: interior mask
    ax2 = axes[1, idx]
    im2 = ax2.imshow(interior, cmap='gray')
    ax2.set_title(f"{bid} - Interior Mask")
    ax2.axis('off')


plt.tight_layout()
plt.show()
```
